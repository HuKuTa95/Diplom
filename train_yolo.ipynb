```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8m.pt')  # Путь к конфигурации модели YOLOv8
results = model.train(
        data='/content/dataset/data.yaml',  # Путь к вашему датасету
        imgsz=640,
        epochs=150,
        close_mosaic=20,
        augment=True,
        batch=6,
        lr0=0.010331723789277684,
        momentum=0.801390767520907,
        weight_decay=0.00023886735267372364,
        dropout=0.20818753926777894,
        optimizer='SGD',
        hsv_s=0.48800351982482415,
        hsv_h=0.00013181168168090673,
        lrf=0.7939149784108267,
        copy_paste=0.2097345954702607,
        bgr=0.24324465455654237,
        mixup=0.35962114955985747,
        flipud=0.5367158474008589,
        degrees=17.01160931185798,
        shear=7.856406748521197,
        perspective=4.9909907708391727e-05
    )
```
